```python
import sympy as smp
```

```python
# Creating symbol x
x = smp.symbols('x')
x
```

```python
# Creating algebraic function x
p = 7*x**5 - 3*x**4 + 12*x**3 - x**2 + 9*x - 6
# 7x5−3x4+12x3−x2+9x−6
```

```python
# Finding 0 of polynomial
roots = smp.solve(p, x)
roots
```

```python
[complex(r.evalf()) for r in roots]
```

```python
# Finding factors of polynomial
p.factor()
```
